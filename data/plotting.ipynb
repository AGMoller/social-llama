```python
import numpy as np
import json
```

```python
data = json.load(open('similarities.json'))
```

```python
print(data.keys())
print(data['baseline'][0].keys())
```

```python
rows = list(data.keys())
cols = list(data['baseline'][0].keys())

print(f"rows: {rows}")
print(f"cols: {cols}")

probs = np.zeros((len(rows), len(cols)))

for row_idx, row in enumerate(rows):
    for col_idx, col in enumerate(cols):
        probabilities = data[row][0][col]
        avg_prob = np.mean(probabilities)
        probs[row_idx, col_idx] = avg_prob
```

```python
probs.round(3)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(probs, cmap='Blues')

# Set ticks and labels
ax.set_xticks(np.arange(len(cols)))
ax.set_yticks(np.arange(len(rows)))
ax.set_xticklabels(cols)
ax.set_yticklabels(rows)

# Rotate the tick labels and set alignment
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Loop over data dimensions and create text annotations
for i in range(len(rows)):
    for j in range(len(cols)):
        text = ax.text(j, i, probs[i, j].round(3), ha="center", va="center", color="black")

# Set title and colorbar
ax.set_title("Confusion Matrix")
cbar = ax.figure.colorbar(im, ax=ax)

# Put an overall label on y and x axis
plt.xlabel('Avg Predicted Personality')
plt.ylabel('True Personality')

# Show the plot
plt.show()
```
